```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
import seaborn as sns
```

```python
import pandas as pd
import os

def append_csv(input_path, output_path):

    if not os.path.exists(output_path):
        
        df = pd.read_csv(input_path)
        
        df.to_csv(output_path, index=False)
    
    else:
        
        input_df = pd.read_csv(input_path)
        
        output_df = pd.read_csv(output_path)
        
        combined_df = pd.concat([output_df, input_df], ignore_index=True)

        combined_df.to_csv(output_path, index=False)

```

```python
post_files = [f for f in os.listdir("./Post")]
pre_files = [f for f in os.listdir("./Pre")]
```

```python
# Pre
for i in pre_files:
    input_path = "./Pre/" + i
    output_path = "./pre_sheep_asthma"
    append_csv(input_path, output_path)
```

```python
# Post
for i in post_files:
    input_path = "./Post/" + i
    output_path = "./post_sheep_asthma"
    append_csv(input_path, output_path)
```

# Post asthma 

```python
df_post = pd.read_csv('post_sheep_asthma.csv')
df_post
```

```python
# Correlation between features
# Pearson product-moment correlation coefficients

plt.figure(figsize=(15,15))
cm = np.corrcoef(df_post.iloc[:,2:7].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 25}, cmap='viridis',yticklabels=df_post.columns[2:7], xticklabels=df_post.columns[2:7])
hm.set_title("Correlation between features post asthma")
plt.show()

```
