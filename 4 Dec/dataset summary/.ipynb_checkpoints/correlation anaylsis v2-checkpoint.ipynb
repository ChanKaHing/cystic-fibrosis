```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
import os

def Pearson(df, name):
    plt.figure(figsize=(15,15))
    cm = np.corrcoef(df.iloc[:,2:8].values.T)
    sns.set(font_scale=1.5)
    hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 25}, cmap='viridis',yticklabels=df.columns[2:8], xticklabels=df.columns[2:8])
    hm.set_title(name)
    plt.show()

```

```python
def kendall(df, name):
    df_fea = df.iloc[:,2:8]
    kendall_corr_matrix = df_fea.corr(method='kendall')
    plt.figure(figsize=(15, 15))
    sns.heatmap(kendall_corr_matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 25}, cmap='viridis')
    plt.title(name)
    plt.show()
```

```python
def spearman(df, name):
    df_fea = df.iloc[:,2:8]
    spearman_corr_matrix = df_fea.corr(method='spearman')
    plt.figure(figsize=(15, 15))
    sns.heatmap(spearman_corr_matrix, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 25}, cmap='viridis')
    plt.title(name)
    plt.show()
```
